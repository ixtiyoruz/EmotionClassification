# 2.3 Text lengths by each class

```python
def get_stats(df):
    df['length'] = df['text'].apply(len)
    df['word_count'] = df['text'].str.split().apply(len)
```

```python
# getting the lengths
get_stats(train_df)
get_stats(val_df)
get_stats(test_df)
```

percentage of <b>word counts</b> binned into 10.
From the graph it can be inferred that the majority of the texts in the range of 7-30, while 19 lengthed texts occured the most.

```python
def plot_count_chart(df,ax, split='train'):
    counts, binEdges= np.histogram(pd.cut(df['word_count'], 10).apply(lambda x: x.left))
    palette_color = sns.color_palette('bright')
    ax.pie(x=counts, labels=np.int32(binEdges[1:]), colors=palette_color, autopct='%.0f%%')
    ax.set_title(split)    
```

# 3. Apply NLP methods


here we will firstly remove the stop words (the words that are frequent and doest not add meaningfull content such as the, a), then we will lemmatize the words (reduce the words into their

```python
import logging
import spacy
logger = logging.getLogger("spacy")
logger.setLevel(logging.ERROR)
# first time 
# spacy.cli.download("en_core_web_lg")

# #loading the english language small model of spacy
en = spacy.load('en_core_web_lg')
sw_spacy = en.Defaults.stop_words
# the words that are common in all classes
REMOVAL  = ['feel', 'href', 'www', 'img', 'src', 'http']
```

## 3.1 lemmmatize and remove stop words

Here we will lemmatize and add the result to text_lemmatized column. Then, we will remove the stop words and keep the result in text_no_stop_words column

```python
# lemmatize
train_df['text_lemmatized'] = train_df['text'].apply(lambda x: " ".join([token.lemma_ for token in en(x)]))

# remove all the stop words from dataframe
train_df['text_no_stop_words'] = train_df['text_lemmatized'].apply(lambda x: "".join([word +" " for word in x.split(' ') if word.lower() not in sw_spacy and word.lower() not in REMOVAL]))
```

see the results

```python
train_df.head()
```

## 3.2 Frequent occuring n-grams
An n-gram is sequenc of n words in a text. n-grams helps us to analyse combination of words (phrases) to get better insights into the data. We can use CountVectorizer from sklearn for this purpose.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

### top n most occuring phrases base on each class

```python

most_N = 10
for label in labels_dict:
    cv = CountVectorizer(ngram_range=(3,3))
    # getting n gram transformers according to each class
    n_gram = cv.fit_transform(train_df.loc[train_df['label']==label, 'text_no_stop_words'])
    count_values = n_gram.toarray().sum(axis=0)
    ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv.vocabulary_.items()], reverse = True))
    ngram_freq.columns = ["frequency", f"{most_N}-gram"]
    sns.barplot(x=ngram_freq['frequency'][:most_N], y=ngram_freq[f'{most_N}-gram'][:most_N])
    plt.title(f'{labels_dict[label]}')
    plt.show()
```

add most common words in all data

```python
word_counts = Counter()
for message in train_df['text_no_stop_words']:
    if isinstance(message, str):
        words = [word.lower() for word in str(message).split() if word.lower() not in sw_spacy]
        word_counts.update(words)
```

```python
n = 50
print(f"adding most common {n} words to the stop words list so that we can ignore them:")
print(", ".join([word[0] for word in word_counts.most_common()[:50]]))
```

```python
for word in word_counts.most_common()[:50]:
    sw_spacy.add(word[0])
```

### visualize the wordclouds 

```python

fig, axs = plt.subplots(3, 2)
axs = axs.flatten()
for idx, emotion in enumerate(emotions):
    word_counts = Counter()
    
    for message in train_df.loc[train_df.name == emotion, 'text_no_stop_words']:
        if isinstance(message, str):
            words = [word.lower() for word in str(message).split() if word.lower() not in sw_spacy]
            word_counts.update(words)

    wordcloud = WordCloud(background_color='white', width=1000, height=500).generate_from_frequencies(word_counts)
    
    axs[idx].imshow(wordcloud, interpolation='bilinear')
    axs[idx].axis('off')
    axs[idx].set_title(emotion)
plt.show()
```
